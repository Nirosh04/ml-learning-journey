```python
import pandas as pd

df = pd.read_csv("titanic.csv")
df.head()
df.groupby("Pclass")["Age"].mean() # durschnittsalter pro Klasse
df.groupby("Sex")["Survived"].sum() #anzahl der überlebenden pro geschlecht
df.groupby("Pclass")["Fare"].mean() #durchschnittspreis pro Klasse
df.groupby("Survived")["Age"].mean() #durschnittsalter der überlebenden
# Alter fehlende Werte mit Mittelwert füllen
df["Age"] = df["Age"].fillna(df["Age"].mean())

# Überprüfen
df.isnull().sum()
```

```python
import pandas as pd

df = pd.read_csv("titanic.csv")
df.head()
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])#"Embarked" ist keine zahl daher mode()
df.isnull().sum()
```

```python
import pandas as pd
df = pd.read_csv("students.csv")
df.head()
df[df["Punkte"] >= 85] #alle die mehr als 85 punkte haben
df[df["Alter"] <= 20]
df[(df["Punkte"] > 80) & (df["Punkte"] < 90)]
df.groupby("Alter")["Punkte"].max() #maximale punktzhal pro Alter
df["Punkte"].agg(["min", "max", "mean"]) #berechent min, max und mean von alter
df.agg({"Punkte": ["min", "max", "mean"], "Alter": ["mean"]})

```

```python
import pandas as pd
df = pd.read_csv("sales.csv")
print(df.head())
df["Umsatz"].sum()
df.groupby("Kategorie")["Umsatz"].mean()
df["Umsatz"].max()
df["Umsatz"].min()
max_cat = df.groupby("Kategorie")["Umsatz"].sum().idxmax()
print(f"Die Kategorie mit dem höchsten Gesamtumsatz ist: {max_cat}")

```

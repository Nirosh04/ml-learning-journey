```python
import requests
import json

url = "https://housing-api-brh2.onrender.com/predict"

data = {
    "longitude": -122.23,
    "latitude": 37.88,
    "housing_median_age": 41,
    "total_rooms": 880,
    "total_bedrooms": 129,
    "population": 322,
    "households": 126,
    "median_income": 8.3252,
    "rooms_per_household": 7.0,
    "bedrooms_per_room": 0.15,
    "population_per_household": 2.5,
    "ocean_proximity": "NEAR OCEAN"
}

response = requests.post(url, json=data)
```

```python
if response.status_code == 200:
    prediction = response.json()
    print("Vorhergesagter Preis:", prediction["predicted_price"])
    print("Antwortzeit (falls dein API es zurückgibt):", prediction.get("latency_sec"))
else:
    print("Fehler:", response.status_code)
    print(response.text)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Stil anpassen (optional)
sns.set(style="whitegrid")

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, marker='o') #liniendiagramm 
plt.title("Beispiel: Umsatzentwicklung") #titel
plt.xlabel("Monat") #beschriftung der x-achse
plt.ylabel("Umsatz ()") #beschriftung der y-achse
plt.show() #ausgabe

produkte = ["Laptop", "Tablet", "Smartphone"]
umsatz = [1200, 600, 850]

plt.bar(produkte, umsatz, color='skyblue')
plt.title("Umsatz pro Produkt")
plt.xlabel("Produkt")
plt.ylabel("Umsatz ()")
plt.show()



```

```python
punkte = [85, 90, 78, 88, 92, 70, 95, 100, 65, 82]
plt.hist(punkte, bins=5, color='orange', edgecolor='black')
plt.title("Verteilung der Punkte")
plt.xlabel("Punkte")
plt.ylabel("Anzahl")
plt.show()
```

```python
import matplotlib.pyplot as plt
df = pd.read_csv("titanic.csv")

df["Survived"].value_counts().plot(kind="bar", color=["red", "green"])
plt.title("berlebende vs. Nicht-berlebende")
plt.xlabel("berlebt (0 = Nein, 1 = Ja)")
plt.ylabel("Anzahl Passagiere")
plt.show()
```

```python
df.boxplot(column="Age", by="Pclass", grid=False)
plt.title("Alter nach Passagierklasse")
plt.suptitle("")
plt.xlabel("Klasse")
plt.ylabel("Alter")
plt.show()
```

```python
df.groupby("Sex")["Survived"].mean().plot(kind="bar", color=["blue", "pink"])
plt.title("berlebensrate nach Geschlecht")
plt.xlabel("Geschlecht")
plt.ylabel("berlebensrate")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("titanic.csv")

df.groupby("Pclass")["Age"].mean().plot(kind='bar', color=["blue", "pink", "green"])
plt.title("Alter pro Klasse")
plt.xlabel("Klasse")
plt.ylabel("Durchschnittsalter")
plt.show()
```

```python
df = pd.read_csv("sales.csv")
sns.barplot(x="Kategorie", y="Umsatz", data=df, estimator="mean", ci = None)
plt.title("Durchschnittlicher Umsatz pro Kategorie")
plt.show()
```

```python
students = pd.read_csv("students.csv")
sns.scatterplot(x="Alter", y="Punkte", data=students)
plt.title("Alter vs Punkte")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Korrelationsmatrix")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("sales.csv")

top5 = df.nlargest(5, "Umsatz")


sns.barplot(x = "Produkt", y = "Umsatz", data = top5.sort_values("Umsatz", ascending = True), palette = "viridis")
plt.title("Top 5 Produkte ")
plt.show()
```

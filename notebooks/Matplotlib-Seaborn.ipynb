```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: adjust plotting style
sns.set(style="whitegrid")

# Example line chart data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, marker='o')  # line chart
plt.title("Example: Revenue Development")  # title
plt.xlabel("Month")  # label for x-axis
plt.ylabel("Revenue ()")  # label for y-axis
plt.show()  # display chart

# Example bar chart data
products = ["Laptop", "Tablet", "Smartphone"]
revenue = [1200, 600, 850]

plt.bar(products, revenue, color='skyblue')
plt.title("Revenue per Product")
plt.xlabel("Product")
plt.ylabel("Revenue ()")
plt.show()



```

```python
punkte = [85, 90, 78, 88, 92, 70, 95, 100, 65, 82]
plt.hist(punkte, bins=5, color='orange', edgecolor='black')
plt.title("Distribution of points")
plt.xlabel("points")
plt.ylabel("number")
plt.show()
```

```python
import matplotlib.pyplot as plt
df = pd.read_csv("../data/titanic.csv")


df["Survived"].value_counts().plot(kind="bar", color=["red", "green"])
plt.title("Survivors vs. Non-survivors")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of passengers")
plt.show()
```

```python
df.boxplot(column="Age", by="Pclass", grid=False)
plt.title("Alter nach Passagierklasse")
plt.suptitle("")
plt.xlabel("Klasse")
plt.ylabel("Alter")
plt.show()
```

```python
df.groupby("Sex")["Survived"].mean().plot(kind="bar", color=["blue", "pink"])
plt.title("berlebensrate nach Geschlecht")
plt.xlabel("Geschlecht")
plt.ylabel("berlebensrate")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("titanic.csv")

df.groupby("Pclass")["Age"].mean().plot(kind='bar', color=["blue", "pink", "green"])
plt.title("Alter pro Klasse")
plt.xlabel("Klasse")
plt.ylabel("Durchschnittsalter")
plt.show()
```

```python
df = pd.read_csv("sales.csv")
sns.barplot(x="Kategorie", y="Umsatz", data=df, estimator="mean", ci = None)
plt.title("Durchschnittlicher Umsatz pro Kategorie")
plt.show()
```

```python
students = pd.read_csv("students.csv")
sns.scatterplot(x="Alter", y="Punkte", data=students)
plt.title("Alter vs Punkte")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Korrelationsmatrix")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("sales.csv")

top5 = df.nlargest(5, "Umsatz")


sns.barplot(x = "Produkt", y = "Umsatz", data = top5.sort_values("Umsatz", ascending = True), palette = "viridis")
plt.title("Top 5 Produkte ")
plt.show()
```

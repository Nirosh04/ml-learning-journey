```python
import pandas as pd

df = pd.read_csv("../data/titanic.csv")
df.head()

df.groupby("Pclass")["Age"].mean()       # average age per class
df.groupby("Sex")["Survived"].sum()      # number of survivors per gender
df.groupby("Pclass")["Fare"].mean()      # average fare per class
df.groupby("Survived")["Age"].mean()     # average age of survivors

# Fill missing age values with the mean
df["Age"] = df["Age"].fillna(df["Age"].mean())

# Check missing values
df.isnull().sum()

```

```python
import pandas as pd

df.head()

# "Embarked" is not a numeric column, so we fill missing values with the mode
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

df.isnull().sum()
```

```python
import pandas as pd

df = pd.read_csv("../data/sales.csv")
print(df.head())

df["Revenue"].sum()
df.groupby("Category")["Revenue"].mean()
df["Revenue"].max()
df["Revenue"].min()

max_cat = df.groupby("Category")["Revenue"].sum().idxmax()
print(f"The category with the highest total revenue is: {max_cat}")

```
